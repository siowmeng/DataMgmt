#### Question 7: Find all possible one-stop flights between any two destinations (not necessarily with the same carrier, but excluding the returning flights) with a stopover between 1 hour and 2 hours.

To find the stopover flights, a join needs to be performed. The helper function below parses the departure and arrival dates (CRS time is used since it is the scheduled time).

```pyspark
from datetime import datetime, timedelta

def dateDepArr(year, month, day, depHHMM, arrHHMM):
    # Add zero padding to month, day and hhmm
    if len(month) < 2:
        month = '0' + month
    if len(day) < 2:
        day = '0' + day        
    if len(depHHMM) < 4:
        numPadding = 4 - len(depHHMM)
        for i in range(numPadding):
            depHHMM = '0' + depHHMM
    
    if depHHMM == '2400':
        depDate = datetime.strptime(year + month + day + ' 0000', '%Y%m%d %H%M') + timedelta(days = 1)
    else:
        depDate = datetime.strptime(year + month + day + ' ' + depHHMM, '%Y%m%d %H%M')
    
    if len(arrHHMM) < 4:
        numPadding = 4 - len(arrHHMM)
        for i in range(numPadding):
            arrHHMM = '0' + arrHHMM
    elif arrHHMM == '2400':
        arrHHMM = '0000'
    
    # Arriving the next day
    if int(arrHHMM) < int(depHHMM):
        arrDate = datetime.strptime(year + month + day + ' ' + arrHHMM, '%Y%m%d %H%M') + timedelta(days = 1)
    else:
        arrDate = datetime.strptime(year + month + day + ' ' + arrHHMM, '%Y%m%d %H%M')
    
    return depDate, arrDate
```

Using the helper function, the below two functions define the two data formats (one for first flight, the other for second flight).

```pyspark
def firstFlight(line):
    depDateTime, arrDateTime = dateDepArr(line[0], line[1], line[2], line[5], line[7])
    
    # Key = (StopOver Airport, CRSArrDateTime in String), Value = (Origin, CRSDepDateTime, CRSArrDateTime, CarrierCode)
    return ((line[17], arrDateTime.strftime('%Y%m%d-%H')), (line[16], depDateTime, arrDateTime, line[8]))
    
def secondFlight(line, offsetHours = 1):
    depDateTime, arrDateTime = dateDepArr(line[0], line[1], line[2], line[5], line[7])
    
    # Key = (StopOver Airport, CRSArrDateTime in String), Value = (Final Destination, CRSDepDateTime, CRSArrDateTime, CarrierCode)
    return ((line[16], (depDateTime - timedelta(hours = offsetHours)).strftime('%Y%m%d-%H')), (line[17], depDateTime, arrDateTime, line[8]))
```

Then, join the two RDDs to identify flights that are potentially connecting flights (stopover between 1 and 2 hours).

```pyspark
# Key = (StopOver Airport, ArrDateTime at StopOver), Value = (Origin, CRSDepDateTime, CRSArrDateTime, CarrierCode)
#CRSArrTime has an entry "2096", simply remove it away

firstFlightRDD = transits \
    .filter(lambda line: (int(line[5]) % 100 < 60) and (int(line[7]) % 100 < 60)) \
    .map(firstFlight)
```

```pyspark
# Key = (StopOver Airport, ArrDateTime at StopOver), Value = (Final Destination, CRSDepDateTime, CRSArrDateTime, CarrierCode)

# Offset the departure hours by 1 hour earlier, so that it can be joined with firstFlightRDD
secondFlightRDD1 = transits \
    .filter(lambda line: (int(line[5]) % 100 < 60) and (int(line[7]) % 100 < 60)) \
    .map(lambda line: secondFlight(line, offsetHours = 1))

# Offset the departure hours by 2 hours earlier, so that it can be joined with firstFlightRDD
secondFlightRDD2 = transits \
    .filter(lambda line: (int(line[5]) % 100 < 60) and (int(line[7]) % 100 < 60)) \
    .map(lambda line: secondFlight(line, offsetHours = 2))

# Both secondFlightRDD1 and secondFlightRDD2 need to be joined with firstFlightRDD, so union them first
secondFlightRDD = secondFlightRDD1.union(secondFlightRDD2)
```

```pyspark
# Key = (StopOver Airport, ArrDateTime at StopOver), Value = ((Origin, CRSDepDateTime, CRSArrDateTime, CarrierCode), (Final Destination, CRSDepDateTime, CRSArrDateTime, CarrierCode))

# Join the two RDDs using (StopOver airport, DateTime (YYYYMMDD-HH)) key to reduce number of entries
stopoverRDD = firstFlightRDD \
    .join(secondFlightRDD) \
    .cache()
```

Finally, apply filter using the conditions:
1. Origin is not the same as eventual destination (i.e. no return flights)
2. Stopover time is between 1 hour and 2 hours

There are a lot of flight data satisfying the these stopover conditions. To get a holistic view, the total number of possible one-stop flights between two airports is aggregated and displayed below (sorted by number of one-stop flights, in descending order).  

In the results, the key is the (Origin, FinalDestination) and value is the number of one-stop flights between these two places.

```pyspark
stopoverRDD \
    .filter(lambda (k, (v1, v2)): (v1[0] != v2[0]) and (timedelta(hours = 1) <= v2[1] - v1[2] <= timedelta(hours = 2))) \
    .map(lambda (k, (v1, v2)): ((v1[0], v2[0]), 1)) \
    .reduceByKey(lambda a, b: a + b) \
    .sortBy(lambda (k, v): v, ascending = False) \
    .collect()
```

#### Question 8: Assume that a passenger wants to travel from Los Angeles International Airport (airport code: LAX) to Orlando International Airport (airport code: MCO), and then go back to Los Angeles (LAX). They depart LAX no earlier than 5:59 (scheduled time), stay at least 3:01 hours in Orlando and then arrive at LAX no later than 11pm. Based on the "scheduled" times, find which carrier has the most flights satisfying these constraints.

Similar to question 7, start by defining two data formats (one for outbound flight LAX -> MCO, the other for inbound flight MCO -> LAX). Then, apply filter using the stated conditions and perform a join.

```pyspark
def outboundFlight(line):
    depDateTime, arrDateTime = dateDepArr(line[0], line[1], line[2], line[5], line[7])
    
    # Key = (StopOver Airport, CRSArrDateTime in String), Value = (Origin, CRSDepDateTime, CRSArrDateTime, CarrierCode)
    return ((line[17], arrDateTime.strftime('%Y%m%d')), (line[16], depDateTime, arrDateTime, line[8]))
    
def inboundFlight(line):
    depDateTime, arrDateTime = dateDepArr(line[0], line[1], line[2], line[5], line[7])
    
    # Key = (StopOver Airport, CRSDepDateTime in String), Value = (Final Destination, CRSDepDateTime, CRSArrDateTime, CarrierCode)
    return ((line[16], depDateTime.strftime('%Y%m%d')), (line[17], depDateTime, arrDateTime, line[8]))

# Key = ('MCO', ArrDate at MCO), Value = (Origin, CRSDepDateTime, CRSArrDateTime, CarrierCode)
outboundRDD = transits \
    .filter(lambda line: (line[16] == 'LAX') and (line[17] == 'MCO')) \
    .filter(lambda line: (int(line[5]) % 100 < 60) and (int(line[7]) % 100 < 60)) \
    .filter(lambda line: (int(line[5]) >= 559)) \
    .map(outboundFlight)

# Key = ('MCO', DepDate from MCO), Value = (Final Destination, CRSDepDateTime, CRSArrDateTime, CarrierCode)
inboundRDD = transits \
    .filter(lambda line: (line[16] == 'MCO') and (line[17] == 'LAX')) \
    .filter(lambda line: (int(line[5]) % 100 < 60) and (int(line[7]) % 100 < 60)) \
    .filter(lambda line: (int(line[7]) <= 2300)) \
    .map(inboundFlight)
```

```pyspark
# Key = ('MCO', TravelDate), Value = ((Origin, CRSDepDateTime, CRSArrDateTime, CarrierCode), (Final Destination, CRSDepDateTime, CRSArrDateTime, CarrierCode))

# Join using ('MCO', Date (YYYYMMDD)) key to reduce number of entries
returnFlightRDD = outboundRDD \
    .join(inboundRDD) \
    .cache()
```

After having the joined data, perform filtering and count the number of return flights for each airline. Finally, only the airline with highest number of return flights in the whole dataset is displayed.  

In the results, the key is the Airline and the value is the number of flights satisfying the constraints.

```pyspark
returnFlightRDD \
    .filter(lambda (k, (v1, v2)): (v1[3] == v2[3]) and (v1[1].date() == v2[2].date()) and (v2[1] - v1[2] >= timedelta(hours = 3, minutes = 1))) \
    .map(lambda (k, (v1, v2)): (v1[3], 1)) \
    .reduceByKey(lambda a, b: a + b) \
    .top(1, key = lambda (k, v): v)
```
